```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
import itertools

# Load the dataset
df = pd.read_csv("amazon.csv")

# 1. Average rating for each product category
avg_rating = df.groupby("category")["rating"].mean()
print("Average Rating by Category:\n", avg_rating)

# 2. Top rating_count products by category
top_products = df.sort_values(["category", "rating_count"], ascending=[True, False])
print("Top Products by Rating Count:\n", top_products.groupby("category").head(5))

# 3. Distribution of discounted prices vs. actual prices
sns.histplot(df["actual_price"], label="Actual Price", color="blue", kde=True)
sns.histplot(df["discounted_price"], label="Discounted Price", color="red", kde=True)
plt.legend()
plt.title("Distribution of Actual vs Discounted Prices")
plt.show()

# 4. Average discount percentage across categories
df["discount_percent"] = ((df["actual_price"] - df["discounted_price"]) / df["actual_price"]) * 100
avg_discount = df.groupby("category")["discount_percent"].mean()
print("Average Discount Percentage by Category:\n", avg_discount)

# 5. Most popular product names (by rating count)
popular_products = df.groupby("product_name")["rating_count"].sum().sort_values(ascending=False)
print("Most Popular Product Names:\n", popular_products.head(10))

# 6. Most popular product keywords (based on frequency)
keywords = list(itertools.chain.from_iterable(df["product_name"].str.lower().str.split()))
keyword_counts = Counter(keywords)
print("Most Popular Keywords:\n", keyword_counts.most_common(10))

# 7. Most popular product reviews (by rating count)
popular_reviews = df.groupby("reviews")["rating_count"].sum().sort_values(ascending=False)
print("Most Popular Product Reviews:\n", popular_reviews.head(10))

# 8. Correlation between discounted price and rating
correlation = df[["discounted_price", "rating"]].corr()
print("Correlation Between Discounted Price and Rating:\n", correlation)

# 9. Top 5 categories based on the highest ratings
top_categories = df.groupby("category")["rating"].mean().sort_values(ascending=False)
print("Top 5 Categories by Highest Ratings:\n", top_categories.head(5))

# 10. Potential areas for improvement (Basic Insights)
print("Summary Statistics:\n", df.describe())

low_rated_products = df[df["rating"] < 3]
print("Low Rated Products:\n", low_rated_products[["product_name", "category", "rating"]].head())

low_sales_products = df[df["rating_count"] < df["rating_count"].quantile(0.25)]
print("Low Sales Products:\n", low_sales_products[["product_name", "category", "rating_count"]].head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('spotify.csv')

# Check for null values
print("Null Values in Each Column:")
print(df.isnull().sum())

# Check for duplicate rows
print("\nNumber of Duplicate Rows:", df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()

# 1. Distribution of Popularity (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=30, kde=True)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()

# 2. Relationship Between Popularity and Duration (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration (ms)', y='Popularity', data=df)
plt.title('Popularity vs Duration')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.show()

# 3. Artist with the Highest Number of Tracks (Count Plot)
artist_track_counts = df['Artist'].value_counts()
print("\nArtist with the Highest Number of Tracks:")
print(artist_track_counts.head(1))

plt.figure(figsize=(12, 8))
sns.countplot(y='Artist', data=df, order=artist_track_counts.index[:10])  # Top 10 artists
plt.title('Number of Tracks per Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()

# 4. Top 5 Least Popular Tracks
least_popular_tracks = df.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
print("\nTop 5 Least Popular Tracks:")
print(least_popular_tracks)

# 5. Average Popularity Among Top 5 Most Popular Artists
average_popularity = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)
top_5_artists = average_popularity.head(5)
print("\nTop 5 Most Popular Artists (by Average Popularity):")
print(top_5_artists)

# 6. Most Popular Tracks of Top 5 Artists
print("\nMost Popular Tracks of Top 5 Artists:")
for artist in top_5_artists.index:
    most_popular_track = df[df['Artist'] == artist].nlargest(1, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
    print(most_popular_track)

# 7. Pair Plot for Multiple Numerical Variables
sns.pairplot(df[['Popularity', 'Duration (ms)']])
plt.show()

# 8. Duration Variation Across Artists (Box Plot)
plt.figure(figsize=(12, 8))
sns.boxplot(x='Duration (ms)', y='Artist', data=df[df['Artist'].isin(artist_track_counts.index[:10])])  # Top 10 artists
plt.title('Duration Variation Across Artists')
plt.xlabel('Duration (ms)')
plt.ylabel('Artist')
plt.show()

# 9. Popularity Distribution Across Artists (Violin Plot)
plt.figure(figsize=(12, 8))
sns.violinplot(x='Popularity', y='Artist', data=df[df['Artist'].isin(artist_track_counts.index[:10])])  # Top 10 artists
plt.title('Popularity Distribution Across Artists')
plt.xlabel('Popularity')
plt.ylabel('Artist')
plt.show()
```
